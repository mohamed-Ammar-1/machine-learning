```python
# plot the data

plt.figure(figsize=(20,6),dpi=80)
plt.scatter(x,y)
plt.xlabel("engine size")
plt.ylabel("co2")
plt.title("data")

plt.show()
```

```python
# function to calculate cost function 
def compute_cost_fun(a_0, a_1, points):
    total_cost = 0
    N = float(len(points))
    
    for i in range(0, len(points)):
        x = points['ENGINESIZE'][i]
        y = points['CO2EMISSIONS'][i]
        total_cost += (y - (a_1 * x + a_0)) ** 2   #MSE
        
    return total_cost/N
```

```python
def step_gradient(a_0_current, a_1_current, points, learning_rate):
    a_1_gradient = 0
    a_0_gradient = 0
    N = float(len(points))

    #Calculate Gradient
    for i in range(0, len(points)):
        x = points['ENGINESIZE'][i]
        y = points['CO2EMISSIONS'][i]
        a_1_gradient += - (2/N) * x * (y - (a_1_current * x + a_0_current))
        a_0_gradient += - (2/N) * (y - (a_1_current * x + a_0_current))
    
    #Update current a_1 and a_0
    a_1_updated = a_1_current - learning_rate * a_1_gradient
    a_0_updated = a_0_current - learning_rate * a_0_gradient

    #Return updated paraa_1eters
    return a_0_updated, a_1_updated


def gradient_descent_runner(points, starting_a_0, starting_a_1, learning_rate, num_iterations):
    a_0 = starting_a_0
    a_1 = starting_a_1
    cost_graph = []

    #For every iteration, optimize a_0, a_1 and compute its cost
    for i in range(num_iterations):
        cost_graph.append(compute_cost_fun(a_0, a_1, points))
        a_0, a_1 = step_gradient(a_0, a_1,points, learning_rate)

    return [a_0, a_1, cost_graph]
```

```python
#hyperparamters
learning_rate = 0.001
initial_a_0 = 0
initial_a_1 = 0
num_iterations = 150
```

```python
a_0, a_1, cost_graph = gradient_descent_runner(data, initial_a_0, initial_a_1, learning_rate, num_iterations)


print ('Optimized a_0:', a_0)
print ('Optimized a_1:', a_1)

print ('Minimized cost:', compute_cost_fun(a_0, a_1, data))
```

```python
# plot the cost graph

plt.figure(figsize=(10, 6), dpi=80)

plt.plot(cost_graph)
plt.xlabel('No. of iterations')
plt.ylabel('Cost')
plt.title('Cost per iteration')
plt.show()
```

```python
# plot line represet data

plt.figure(figsize=(12, 6), dpi=80)

plt.scatter(x, y)

pred = a_1 * x + a_0


plt.plot(x, pred, c='r')
plt.xlabel('engine size')
plt.ylabel('co2')
plt.title('Line of best fit')
plt.show()
```

```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from mlxtend.evaluate import bias_variance_decomp
```

```python
# read the data
data=pd.read_csv("E:\machine learning\FuelConsumptionCo2.csv")
data.head()
```

```python
# define x and y
x=data[["ENGINESIZE"]].values
y=data["CO2EMISSIONS"].values
```

```python
# train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)

model.coef_
```

```python
 # predicting the accuracy score

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

y_pred=model.predict(x_test)

r_score=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
Rmse=np.sqrt(mean_squared_error(y_test,y_pred))

print("r2_score is :",r_score)
print("mse is :",mse)
print("Rmse is :",Rmse)

a_0=model.intercept_
a_1=model.coef_

print("a_0 is :",a_0)
print("a_1 is :",a_1)

```

```python
# Get Bias and Variance - bias_variance_decomp function

avg_expected_loss,avg_bias,avg_variance=bias_variance_decomp(model,x_train,y_train,x_test,y_test,loss="mse",num_rounds=200,random_seed=1)

print(f"average expected loss is : {round(avg_expected_loss,4)}")
print(f"average bias is : {round(avg_bias,4)}")
print(f"average variance is : {round(avg_variance,4)}")
```
